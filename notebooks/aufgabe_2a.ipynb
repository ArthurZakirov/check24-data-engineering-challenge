### Welche Provider haben die höchsten Verkaufszahlen? (Stückzahl)

```python
verkaufszahlen = df["providerName"].value_counts()
verkaufszahlen.plot(kind='bar', ylabel='Verkaufszahl', title='Verkaufszahlen nach Provider')
plt.show()

tarife = df["tariffName"].value_counts()[:10]
tarife.plot(kind='bar', ylabel='Anzahl', title='Tarife nach Anzahl an Verkäufen')
plt.show()
```
